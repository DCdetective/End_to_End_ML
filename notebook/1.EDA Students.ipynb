```python
data['total score'] = data['math score'] + data["writing score"] + data["reading score"]
```

```python
data['average'] = data['total score']/3
data.head()
```

```python
math_full = data[ data['math score']==100]['average'].count()
reading_full = data[data['reading score']==100]['average'].count()
writing_full = data[data['writing score']==100]['average'].count()

print("The no of students got full in math : ",math_full)
print("The no of students got full in reading : ",reading_full)
print("The no of students got full in writing: ",writing_full)
```

```python
math_fail = data[data['math score']<=20]['average'].count()
reading_fail =  data[data['reading score']<=20]['average'].count()
writing_fail =  data[data['writing score']<=20]['average'].count()

print("The no of students got fail in math: ",math_fail)
print("The no of students got fail in reading: ",reading_fail)
print("The no of students got fail in writing: ",writing_fail)
```
